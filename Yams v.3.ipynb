```python
import random
import math

conserve = []
score = []
tot_total = []
N=5

print("Les dés sont numérotés de 1 à %d en partant de la gauche."%(N))


while len(score) < 13:
    
    # Premier lancer à N dés

    roll1 = [0]*N
    for i in range(N):
        roll1[i] = random.randrange(1,7)
    print("Premier lancer :",roll1)

    dices_conserve = input("Quels dés voulez-vous garder (les séparer par des blancs) ? ")

    # Deuxième lancer avec conservation de certains dés du premier lancer + classement de la liste

    index_reroll = dices_conserve.split()
    roll2 = []
    for i in index_reroll:
        roll2.append((roll1[int(i)-1]))
    roll2 += [random.randrange(1,7) for i in range(N-len(roll2))]
    roll2 = sorted(roll2)
    print("Deuxième lancer :",roll2)    

    # Création d'une liste avec comptage des occurrences de chaque élément du deuxième lancer

    i = 1
    check=[]
    for i in roll2:
        check.append(roll2.count(i))

    # Création d'un dictionnaire de stockage des différents scores possibles du lancer 2

    value = {}

    # comprends pas la suite !!!

    tot = [0]*7
    for i in range(1,7):
        tot[i] = roll2.count(i)*i
    for i in range(1,7):
        value['Total de %d' % i] = tot[i]  # oublié le pourquoi des signes cabalistiques

    # comptage des différents scores possibles du lancer 2

    # Brelan
    if 3 in check:
        nb = check.index(3)
        value["Brelan"] = roll2[nb]*3
    elif 4 in check:
        nb = check.index(4)
        value["Brelan"] = roll2[nb]*3
    elif 5 in check:
        nb = check.index(5)
        value["Brelan"] = roll2[nb]*3
    else:
        value["Brelan"] = 0

    # Carré
    if 4 in check:
        nb = check.index(4)
        value["Carré"] = roll2[nb]*4
    elif 5 in check:
        nb = check.index(5)
        value["Carré"] = roll2[nb]*4
    else:
        value["Carré"] = 0

    # Full
    if 3 in check and 2 in check:
        value["Full"] = 25
    else:
        value["Full"] = 0

    # Petite suite
    if 1 in roll2 and 2 in roll2 and 3 in roll2 and 4 in roll2:
        value["Petite suite"] = 30
    elif 2 in roll2 and 3 in roll2 and 4 in roll2 and 5 in roll2:
        value["Petite suite"] = 30
    elif 3 in roll2 and 4 in roll2 and 5 in roll2 and 6 in roll2:
        value["Petite suite"] = 30
    else:
        value["Petite suite"] = 0

    # Grande suite
    if 1 in roll2 and 2 in roll2 and 3 in roll2 and 4 in roll2 and 5 in roll2:
        value["Grande suite"] = 40
    elif 2 in roll2 and 3 in roll2 and 4 in roll2 and 5 in roll2 and 6 in roll2:
        value["Grande suite"] = 40
    else:
        value["Grande suite"] = 0

    # Yams
    if 5 in check:
        value["Yams"] = 50
    else:
        value["Yams"] = 0

    # Chance
    value["Chance"] = sum(roll2)   

    # Affichage propre des différentes combinaisons

    combinations = value.keys()
    for comb in combinations:
        if comb not in conserve:
            print(comb,":", value[comb])

    # Récupération du nom (pour empêcher la réutilisation) et du score (pour l'ajouter au total)

    question_cons = input("Quel score gardez-vous ? ")
    if question_cons not in conserve:
        sc = None
        if question_cons[:9] == "Total de ":
            i = question_cons[-1]
            sc = tot[int(i)]
        elif question_cons == "Brelan":
            sc = value["Brelan"]
        elif question_cons == "Carré":
            sc = value["Carré"]
        elif question_cons == "Full":
            sc = value["Full"]
        elif question_cons == "Petite suite":
            sc = value["Petite suite"]
        elif question_cons == "Grande suite":
            sc = value["Grande suite"]
        elif question_cons == "Yams":
            sc = value["Yams"]
        elif question_cons == "Chance":
            sc = value["Chance"]
        else:
            print("Réponse non-valide !")
        if sc != None:
            score.append(sc)
            conserve.append(question_cons)
        if sc != None and question_cons[:9] == "Total de ":
            tot_total.append(sc)
    else:
        print("Combinaison déja utilisée !")

    print("Votre total est de", sum(score),"!")

final_score = sum(score)    
if sum(tot_total) >= 63:
         final_score += 35

print("Partie finie ! Votre score final est de", final_score,"!")
```
